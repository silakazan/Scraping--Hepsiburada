```python
liste = []
a = 1
while a <= 50:
    url = requests.get("https://www.hepsiburada.com/ara?q=tablet&sayfa=" + str(a) +"", headers=header)
    soup = BeautifulSoup(url.content,"lxml")
    st1=soup.find("div",attrs={"class":"contain-lg-3 contain-md-3 contain-sm-3 fluid with-bottom-border"})
    st2 = st1.find("ul",attrs={"data-bind" :"css: {grid: isGridSelected, list: isListSelected}"})
    st3 = st2.find_all("li",attrs={"data-index":"1"})





    for detaylar in st3:
        link_sonu= detaylar.a.get("href")
        link_basi= "https://www.hepsiburada.com/"
        link= link_basi + link_sonu
        r1= requests.get(link,headers = header)
        soup1=BeautifulSoup(r1.content,"lxml")
        
        
        try:
            ürün_ismi=soup1.find("h1",attrs={"itemprop":"name"}).text.strip()
            
        except:
            ürün_ismi = "None"
        
        try:
            ürün_marka=soup1.find("span",attrs={"class":"brand-name"}).text.strip()
        except:
            ürün_marka= "None"
        
        try:
            indirimli_fiyat=soup1.find("span",attrs={"itemprop":"price"}).text.strip().replace("\n"," ")
        except:
            indirimli_fiyat= "None"
            
        try:
            orjinal_fiyat=soup1.find("del",attrs={"id":"originalPrice"}).text.strip().replace("\n"," ")
        except:
            orjinal_fiyat= "None"
            
        try:
            indirim=soup1.find("span",attrs={"id":"product-discount-rate"}).text.strip().replace("indirim","").strip()
        except:
            indirim = "None"
            
        try:            
            puan=soup1.find("span",attrs={"class":"rating-star"}).text.strip()
        except:
            puan= "None"
            
        try:
            degerlendirme=soup1.find("div",attrs={"id":"comments-container"}).text.strip().replace("Değerlendirme" , "")
        except:
            degerlendirme = "None"

        try:
            sepette_indirim=soup1.find("span",attrs={"class":"extra-discount-size"}).text.strip().replace("\n"," ").replace("SEPETTE","")
        except:
            sepette_indirim= "None"
            
        try:
            sepette_indirimli_fiyat=soup1.find("div",attrs={"class":"extra-discount-price"}).text.strip()
        except:
            sepette_indirimli_fiyat= "None"


        liste.append([ürün_ismi,ürün_marka,orjinal_fiyat,indirim,indirimli_fiyat,sepette_indirim,sepette_indirimli_fiyat,puan,degerlendirme,link])
    
                 
    a = a + 1
   







```

```python
pd.DataFrame(liste)
```

```python
df= pd.DataFrame(liste)
df.columns=["Ürün_adı","Marka","orjinal_fiyat","İndirim_oranı","indirimli_fiyat","sepette_indirim_oranı","sepette_indirimli_fiyat","Puan","Değerlendirme","Link"]
df
```
